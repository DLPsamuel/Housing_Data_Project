```python
def get_census_acs_2017():
    """
    Fetches the Total Population (B01003_001E) for all states 
    from the 2017 ACS 5-Year Data API.
    """
    # 1. Base URL for ACS 5-Year 2017
    base_url = "https://api.census.gov/data/2017/acs/acs5"

    # 2. Define the parameters
    # 'get': The variables to retrieve. 
    #        NAME = Geographic name (e.g., "Alabama")
    #        B01003_001E = Total Population Estimate
    # 'for': The geography level. 'state:*' means all states.
    params = {
        "get": "NAME,B01003_001E",
        "for": "state:*"
    }

    # NOTE: If you have an API Key, add it to params:
    params['key'] = API_KEY

    print(f"--- Requesting 2017 ACS Data from {base_url} ---")

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()

        # The Census API returns a list of lists (like a CSV).
        # The first row is the header.
        data = response.json()
        
        headers = data[0]
        rows = data[1:]

        print(f"Successfully retrieved data for {len(rows)} states.\n")
        
        # Print header
        print(f"{headers[0]:<20} | {headers[1]}")
        print("-" * 35)

        # Print first 5 rows as a preview
        for row in rows[:5]:
            name = row[0]
            population = row[1]
            print(f"{name:<20} | {population}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching Census data: {e}")

if __name__ == "__main__":
    get_census_acs_2017()
```
